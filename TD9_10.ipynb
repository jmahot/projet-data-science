# Import sources

```python
import pandas as pd
import numpy as np

df=pd.read_csv("valeursfoncieres-2023.txt",sep="|", low_memory=False)
```

# Préparation des données 

```python
# supprimer les colonnes où il y a que des NaN
df = df.dropna(axis = 1, how = 'all')
supprimer = ['No disposition', 'No voie', 'B/T/Q', 'Type de voie', 'Code voie', 'Voie', 'No plan', 'Prefixe de section', 'Nature culture speciale', 'No Volume', '1er lot', '2eme lot', '3eme lot', '4eme lot', '5eme lot']
df = df.drop(columns=supprimer)
```

```python
df.columns
```

```python
# Concentration sur la région de l'ile-de-france
departements_idf = ['75', '77', '78', '91', '92', '93', '94', '95']
df_idf = df[df['Code departement'].astype(str).isin(departements_idf)]
print(df_idf.head(10))
```
